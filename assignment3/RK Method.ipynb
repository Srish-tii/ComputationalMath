```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def R(z):
    return 1 + z + z**2/2
```

```python
real = np.linspace(-3, 1, 400)
imaginary = np.linspace(-3, 3, 400)
r, i = np.meshgrid(real, imaginary)
z = r + 1j * i
magnitude = np.abs(R(z))
```

```python
plt.figure(figsize=(8, 6))
plt.contourf(r, i, magnitude, levels=[0, 1], colors=['orange'], alpha=0.3)
plt.contour(r, i, magnitude, levels=[1], colors='orange', linewidths=1)
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.title('Region of Absolute Stability')
plt.grid(True)
plt.show()
```
