```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def R(z):
    return (12 + 5*z)/((z-3)*(z-4))
```

```python
real = np.linspace(-15, 15, 400)
imaginary = np.linspace(-10,10, 400)
r, i = np.meshgrid(real, imaginary)
z = r + 1j * i
magnitude = np.abs(R(z))
plt.figure(figsize=(8, 6))
plt.contourf(r, i, magnitude, levels=[0, 1], colors=['orange'], alpha=0.3)
plt.contour(r, i, magnitude, levels=[1], colors='orange', linewidths=1)
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.title('Region of Absolute Stability')
plt.grid(True)
plt.show()
```
